```python
%matplotlib inline
```

```python
# import dependencies.
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept","Oct", "Nov", "Dec"]
    
# set y_axis to a list of floats as the total fare in US dollars accumlulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)--matlab
```

```python
# Create the plot with ax.plt()---object oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Cerate the plt with ax.plt---object oriented method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()---object oriented method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create th plot--plt.show() should only be used once per session bc it CAN cause the graph to display improperly--matlab
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a marker, aline color, a line width, and a label for the legend--matlab
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label = "Boston")

# Create label for the x a nd y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit betweeb 0 and 45
plt.ylim(0,45)

# Create title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot and add a marker, aline color, a line width, and a label for the legend--matlab
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label = "Boston")

# Create label for the x a nd y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit betweeb 0 and 45
plt.ylim(0,45)

# Create title
plt.title("PyBer Fare by Month")

# Create grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# create the plot--matlab
plt.bar(x_axis, y_axis)
```

```python
# create the plot--matlab
plt.bar(x_axis, y_axis, color="green", label= "Boston")

# great labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create a title
plt.title("PyBer Fare by Month")

# add the legend
plt.legend()
```

```python
# create the plot---horizontal--matlab
plt.barh(x_axis,y_axis)
```

```python
# create the plot===horizontal but switch axes--matlab
plt.barh(y_axis, x_axis)
```

```python
# creat the plot---horizontal but reorder y_axis by using gca().invert_yaxis()--get current axes--matlab
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# creat the plot---horizontal but reorder y_axis by using gca().invert_yaxis()--get current axes--matlab
plt.barh(x_axis, y_axis, color="magenta", label ="Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# create the plot with ax.plt---object oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# create the plt with as.plt()---object oriented method
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# create the plt with as.plt()---object oriented method--swith axes
fig, ax=plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# create the plt with as.plt()---object oriented method
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
```

```python
# scatter with Matlab
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, c="red", label= "Chicago")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.xlim(0,50)
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# scatter with object oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#skills drill
# size*5, color sky blue, color 20% transparent
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,
           s=[i*5 for i in y_axis],
           alpha = 0.8,
           facecolor = "skyblue",
           edgecolor = "black",
           linewidths = 2,
           label = "Boston"
          )
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.legend()
```

```python
# Using Matlab
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,
        labels = x_axis,
        explode = explode_values,
        colors = colors,
        autopct = "%1.1f%%")
plt.show()
```

```python
# object oriented method--skills drill
explode_chart = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize = (8,8))
ax.axis("equal")
ax.pie(y_axis,
       labels = x_axis,
       explode = explode_chart,
       autopct = "%1.1f%%",
       counterclock = False,
       startangle = 95,
       colors = colors,
       shadow = True,
      )
plt.show()
```
